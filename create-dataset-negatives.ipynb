```python
import polars as pl
```

```python
t = pl.scan_parquet("./train-with-negatives.parquet")
```

```python
t1 = pl.scan_parquet("./train-with-negative-docs-1.parquet")
t2 = pl.scan_parquet("./train-with-negative-docs-2.parquet")
t3 = pl.scan_parquet("./train-with-negative-docs-3.parquet")
t4 = pl.scan_parquet("./train-with-negative-docs-4.parquet")
t5 = pl.scan_parquet("./train-with-negative-docs-5.parquet")
t6 = pl.scan_parquet("./train-with-negative-docs-6.parquet")
t95 = pl.scan_parquet("./train-with-negative-docs-95.parquet")
t96 = pl.scan_parquet("./train-with-negative-docs-96.parquet")
t97 = pl.scan_parquet("./train-with-negative-docs-97.parquet")
t98 = pl.scan_parquet("./train-with-negative-docs-98.parquet")
t99 = pl.scan_parquet("./train-with-negative-docs-99.parquet")
t100 = pl.scan_parquet("./train-with-negative-docs-100.parquet")
```

```python
t = t.with_columns(pl.col("body_length_norm").fill_null(0))
t1 = t1.with_columns(pl.col("body_length_norm").fill_null(0))
t2 = t2.with_columns(pl.col("body_length_norm").fill_null(0))
t3 = t3.with_columns(pl.col("body_length_norm").fill_null(0))
t4 = t4.with_columns(pl.col("body_length_norm").fill_null(0))
t5 = t5.with_columns(pl.col("body_length_norm").fill_null(0))
t6 = t6.with_columns(pl.col("body_length_norm").fill_null(0))
t95 = t95.with_columns(pl.col("body_length_norm").fill_null(0))
t96 = t96.with_columns(pl.col("body_length_norm").fill_null(0))
t97 = t97.with_columns(pl.col("body_length_norm").fill_null(0))
t98 = t98.with_columns(pl.col("body_length_norm").fill_null(0))
t99 = t99.with_columns(pl.col("body_length_norm").fill_null(0))
t100 = t100.with_columns(pl.col("body_length_norm").fill_null(0))

t = t.with_columns(pl.col("title_length_norm").fill_null(0))
t1 = t1.with_columns(pl.col("title_length_norm").fill_null(0))
t2 = t2.with_columns(pl.col("title_length_norm").fill_null(0))
t3 = t3.with_columns(pl.col("title_length_norm").fill_null(0))
t4 = t4.with_columns(pl.col("title_length_norm").fill_null(0))
t5 = t5.with_columns(pl.col("title_length_norm").fill_null(0))
t6 = t6.with_columns(pl.col("title_length_norm").fill_null(0))
t95 = t95.with_columns(pl.col("title_length_norm").fill_null(0))
t96 = t96.with_columns(pl.col("title_length_norm").fill_null(0))
t97 = t97.with_columns(pl.col("title_length_norm").fill_null(0))
t98 = t98.with_columns(pl.col("title_length_norm").fill_null(0))
t99 = t99.with_columns(pl.col("title_length_norm").fill_null(0))
t100 = t100.with_columns(pl.col("title_length_norm").fill_null(0))
```

```python
t_full = t.join(t1, on="index", suffix="_1").join(t2, on="index", suffix="_2").join(
    t3, on="index", suffix="_3"
).join(t4, on="index", suffix="_4").join(t5, on="index", suffix="_5").join(
    t6, on="index", suffix="_6"
).join(
    t95, on="index", suffix="_95"
).join(
    t96, on="index", suffix="_96"
).join(
    t97, on="index", suffix="_97"
).join(
    t98, on="index", suffix="_98"
).join(
    t99, on="index", suffix="_99"
).join(
    t100, on="index", suffix="_100"
).collect()
```

```python
t_full = t_full.sample(fraction=1.0, shuffle=True, seed=42)
n = len(t_full)
train_end = int(0.8 * n)
val_end = int(0.9 * n)
train_set = t_full[:train_end]
val_set = t_full[train_end:val_end]
test_set = t_full[val_end:]
```

```python
train_set.write_parquet("./search_engine/ranking-model/data/train-split.parquet")
val_set.write_parquet("./search_engine/ranking-model/data/val-split.parquet")
test_set.write_parquet("./search_engine/ranking-model/data/test-split.parquet")
```
